<h1>Geopandas</h1>

GeoPandas is particularly relevant given that we are using geospatial data in Lab 2, Lab 3 and Assignment 1. GeoPandas extends the datatypes used by pandas to allow spatial operations on geometric types. You will notice that this is quite convinient to, for example, vizualise how different countries fare regarding a certain metric: you just need to add a Series with data on the polygons defining each country and plot the geospatial data.

```python
import geopandas as gpd
import pandas as pd

# below we will read data already included in GeoPandas, with info about countries
# notice the column geometry that includes the polygon of each country

world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world.head()
```

```python
# do a plot representing the population size of each country
world.plot("pop_est", figsize=(15,10))
```

```python
# Create a (Geo)DataFrame based on imported data
happinessdataframe = pd.read_excel('DataPanelWHR2021C2.xls', index_col=[0,1]) 
happinessdataframe2018 = happinessdataframe.loc(axis=0)[:,2018]
world_happiness_final = world.merge(happinessdataframe2018, how="left", left_on=['name'], right_on=['Country name'])
world_happiness_final.head()
```

```python
import matplotlib.pyplot as plt

world_happiness_final.plot("Log GDP per capita", 
                           figsize=(15,10),
                           legend=True,
                           edgecolor="white",
                           legend_kwds={"label":"Log GDP per capita", "orientation":"horizontal"})
```

<h1>Seaborn</h1>

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. 

Essentially, Seaborn can be used to, with a couple of lines, change the parameters of matplotlib in such a way that vizualisation look good. Seaborn already includes a set of pre-defined styles and themes. Below we provide some examples of how to use Seaborn. Feel free to explore.

In the following link you'll be able to find extra examples: https://seaborn.pydata.org/examples/index.html

```python
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.arange(0,10,0.1)             # create some data
y = np.cos(np.arange(0,10,0.1))     # create some data
```

```python
fig, ax = plt.subplots()   # create a figure and a set of subplots

ax.plot(x, np.sin(x))   # plot scatter plot in the corresponding Axes (ax)
ax.plot(x, np.cos(x))   # plot scatter plot in the same Axes (ax)

plt.legend(['sin','cos'])  # add a legend
plt.show()                 # display all open figures
```

```python
sns.set_style("darkgrid")
sns.set_context("talk")

fig, ax = plt.subplots()   # create a figure and a set of subplots

ax.plot(x, np.sin(x))   # plot scatter plot in the corresponding Axes (ax)
ax.plot(x, np.cos(x))   # plot scatter plot in the same Axes (ax)

plt.legend(['sin','cos'])  # add a legend
plt.show()                 # display all open figures
```

```python
vals1 = np.random.standard_normal(1000)
vals2 = np.random.standard_normal(1000)

df = pd.DataFrame({'one' : vals1, 'two' : vals2, 'sum_positive' : vals1+vals2 > 0})
```

```python
sns.set_style("white")
sns.set_context("paper")
```

```python
sns.jointplot(data=df, x='one', y='two', hue='sum_positive')
```

If you do not reset the matplotlib parameters altered by seaborn, you'll notice that the changes will also affect previous plots, if you run them again (e.g., try to run the Geopandas cells above). The next line resets the parameters set by seaborn.

```python
sns.reset_orig()
```
